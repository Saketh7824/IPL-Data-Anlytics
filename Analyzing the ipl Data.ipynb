```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
```

```python
data=pd.read_csv('C:/Data Sets/matches.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data['date']=pd.to_datetime(data['date'])
```

```python
data['winner'].value_counts()
```

```python
#Getting the frequency of most man of the match awards
data['player_of_match'].value_counts()#[:10]
```

```python
#Getting the top 10 players with most man of the match awards
data['player_of_match'].value_counts()[:10]
```

```python
# making a bar-plot for top 10 players with most man of the match awards
plt.figure(figsize=(8,5))
sns.barplot(data['player_of_match'].value_counts()[:10].index,data['player_of_match'].value_counts()[:10].values)
plt.xticks(rotation=75)
plt.show()
```

```python
data['result'].value_counts()
```

```python
data['toss_winner'].value_counts()
```

```python
batting_first=data[data['win_by_runs']!=0]
batting_first.head()
```

```python
plt.figure(figsize=(8,8))
plt.title('Distribution of Runs',size=20)
sns.histplot(batting_first['win_by_runs'])
plt.show()
```

```python
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(8,8))
plt.title('Teams with most wins after batting first',size=20)
plt.xlabel('Teams',size=15)
plt.ylabel('Count',size=15)
sns.barplot(batting_first['winner'].value_counts().index,batting_first['winner'].value_counts().values)
plt.xticks(rotation=55)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.title('Teams with most wins after batting first',size=20)
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
batting_second=data[data['win_by_wickets']!=0]
batting_second
```

```python
plt.figure(figsize=(8,8))
sns.histplot(batting_second['win_by_wickets'],bins=30)
plt.show()
```

```python
batting_second['winner'].value_counts()
```

```python
plt.figure(figsize=(8,8))
plt.title('Teams with most wins after batting second',size=20)
plt.xlabel('Teams',size=15)
plt.ylabel('Count',size=15)
sns.barplot(batting_second['winner'].value_counts().index,batting_second['winner'].value_counts().values,linewidth=1,edgecolor='black')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%',startangle=90)
plt.show()
```
